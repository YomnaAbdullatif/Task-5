```python
from sklearn.model_selection import train_test_split

y = df['casualty_severity'].values.reshape(-1,)
X = df.drop(['casualty_severity','lsoa_of_casualty','collision_index','collision_reference'], axis=1)
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.5, random_state=0, stratify=y)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

**Random forest Algorthim**

**bold text**

```python
from sklearn.ensemble import RandomForestClassifier  # For classification
from sklearn.ensemble import RandomForestRegressor  # For regression
from sklearn.metrics import accuracy_score
# Create a Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)  # Adjust parameters as needed

# Train the model
rf_model.fit(X_train, y_train)

# Make predictions
y_pred = rf_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
feature_importance = rf_model.feature_importances_
print(feature_importance)
feature_names = X_train.columns
for feature, importance in zip(feature_names, feature_importance):
    print(f"{feature}: {importance:.3f}")
```

**SVM**

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Create an SVM model
svm_model = SVC(kernel='rbf')

# Train the model
svm_model.fit(X_train, y_train)

# Make predictions
y_pred = svm_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

**RNN**

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
```

```python
rnn_model = Sequential()
rnn_model.add(SimpleRNN(units=32, input_shape=(1, 15)))  # Adjust units and input_shape as needed
rnn_model.add(Dense(1))

# Compile the model
rnn_model.compile(optimizer='adam', loss='mean_squared_error')

X = np.array(X_train)
y = np.array(y_train)

# Reshape the input data to be 3-dimensional
X = X.reshape(X.shape[0], 1, X.shape[1])

# Train the model
rnn_model.fit(X, y, epochs=10, batch_size=1)
```

```python
X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[2])
y_pred = rnn_model.predict(X_test)
y_pred = np.round(y_pred)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

**Lineear Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score # Import appropriate metrics for regression

reg_model = LinearRegression()

# Reshape X back to 2 dimensions
X_2d = X.reshape(X.shape[0], X.shape[2])

reg_model.fit(X_2d, y) # Use the reshaped X
y_pred = reg_model.predict(X_2d) # Use the reshaped X

# Use metrics suitable for regression problems
mse = mean_squared_error(y_train, y_pred)
r2 = r2_score(y_train, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

**Gradient Boosting**

```python
from sklearn.ensemble import GradientBoostingClassifier

# X already has the correct shape (24658, 15)
#model = X.reshape(X.shape[0], X.shape[2])
grd_model = GradientBoostingClassifier()
grd_model.fit(X,y)
y_pred = grd_model.predict(X)

# Evaluate the model
accuracy = accuracy_score(y, y_pred)
print("Accuracy:", accuracy)
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
from sklearn.metrics import confusion_matrix
X_test = X_test.reshape(X_test.shape[0], X_test.shape[2])
y_pred = grd_model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
